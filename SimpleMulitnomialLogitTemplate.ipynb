```python
import io
import pandas as pd
file_name = 'predictheartdisease copy.xlsx'
df = pd.read_excel(file_name)
```

```python
#glucose is null, exclude and step wise later
df.isnull().sum()
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
import statsmodels.api as sm
```

```python
chd_unique = df['TenYearCHD'].unique()
chd_unique
```

```python
X = df.drop(columns = 'TenYearCHD')
X = sm.add_constant(X)
y = df['TenYearCHD']
```

```python
model = sm.MNLogit(y,X)
result = model.fit()

result.summary()
```

```python
predicted_probs = result.predict(X)

# Getting the predicted category with the highest probability
predicted_category = predicted_probs.idxmax(axis=1)

# Creating a confusion matrix
confusion_matrix = pd.crosstab(y, predicted_category, rownames=['Actual'], colnames=['Predicted'], margins=True)

# Displaying the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix)

# Calculating classification accuracy
correct_predictions = confusion_matrix.loc['All', 'All']
total_predictions = confusion_matrix.loc['All', 'All']
accuracy = correct_predictions / total_predictions

print(f"\nClassification Accuracy: {accuracy * 100:.2f}%")
```
